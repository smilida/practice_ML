```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import mglearn
import sklearn
from sklearn.model_selection import train_test_split
```

```python
mglearn.plots.plot_scaling()
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data,
                                                    cancer.target,
                                                    random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
# 
X_train_scaled = scaler.transform(X_train)
# 
print("transformed shape:{}".format(X_train_scaled.shape))
print("per-featured minimum before scaling:\n{}".format(X_train.min(axis=0)))
print("per-featured maximum before scaling:\n{}".format(X_train.max(axis=0)))
print("per-featured minimum after scaling:\n{}".format(X_train_scaled.min(axis=0)))
print("per-featured maximum after scaling:\n{}".format(X_train_scaled.max(axis=0)))
```

```python
# 
X_test_scaled = scaler.transform(X_test)
# 
print("per-featured minimum after scaling:\n{}".format(X_test_scaled.min(axis=0)))
print("per-featured maximum after scaling:\n{}".format(X_test_scaled.max(axis=0)))
```

```python
from sklearn.datasets import make_blobs
# 
X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
# 
X_train, X_test = train_test_split(X, random_state=5, test_size=.1)

# 
fig, axes = plt.subplots(1, 3, figsize=(13, 4))
axes[0].scatter(X_train[:, 0], X_train[:, 1],
                c=mglearn.cm2(0), label="Training set", s=60)
axes[0].scatter(X_test[:, 0], X_test[:, 1], marker='^',
                c=mglearn.cm2(1), label="Test set", s=60)
axes[0].legend(loc='upper left')
axes[0].set_title("Original Data")

# MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 
axes[1].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
                c=mglearn.cm2(0), label="Training set", s=60)
axes[1].scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], marker='^',
                c=mglearn.cm2(1), label="Training set", s=60)
axes[1].set_title("Scaled Data")

# 
# 01
# 
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)

# 
axes[2].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
                c=mglearn.cm2(0), label="Training set", s=60)
axes[2].scatter(X_test_scaled_badly[:, 0], X_test_scaled_badly[:, 1], marker='^',
                c=mglearn.cm2(1), label="Training set", s=60)
axes[2].set_title("Improperly Scaled Data")

for ax in axes:
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
```

```python
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(cancer.data,
                                                    cancer.target,
                                                    random_state=0)
svm = SVC(C=100)
svm.fit(X_train, y_train)
print("Test set accuracy:{:.2f}".format(svm.score(X_test, y_test)))
```

```python
# 0-1
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# SVM
svm.fit(X_train_scaled, y_train)

# 
print("Scaled test set accuracy:{:.2f}".format(svm.score(X_test_scaled, y_test)))
```

```python
# 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# SVM
svm.fit(X_train_scaled, y_train)

# 
print("Scaled test set accuracy:{:.2f}".format(svm.score(X_test_scaled, y_test)))
```
